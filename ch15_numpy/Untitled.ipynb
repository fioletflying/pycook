```python
# 随机漫步图

from random import choice

class RandomWalk():
    """随机漫步的类"""
    
    def __init__(self,num_points =5000):
        self.num_points = num_points
        
        self.x_values = [0]
        self.y_values = [0]
        
        
    def fill_walk(self):
        # 生成需要行走的方向
        while len(self.x_values) < self.num_points:
            x_direction = choice([1,-1])  # x的方向的确定
            x_distance = choice([0,1,2,3,4]) # x的值的确定
            x_step = x_direction * x_distance # x行走的确定
            
            y_direction = choice([1,-1]) # y的方向的确定
            y_distance = choice([0,1,2,3,4]) # y的值的确定
            y_step = y_direction * y_distance # y行走的确定
            
            if x_step ==0 and y_step == 0:
                continue
                
            next_x = self.x_values[-1] + x_step # 下一个点x值
            next_y = self.y_values[-1] + y_step # 下一个点y值
            
            self.x_values.append(next_x)
            self.y_values.append(next_y)
            
    
```

```python
import matplotlib.pyplot as plt

while True:
    
    rw = RandomWalk()
    rw.fill_walk()
    
    point_numbers = list(range(rw.num_points))
    plt.scatter(rw.x_values,rw.y_values,s=15,c=point_numbers,cmap=plt.cm.Blues)
    # 突出显示起点和终点
    plt.scatter(0,0,c='green',edgecolors='none',s=100)
    plt.scatter(rw.x_values[-1],rw.y_values[-1],c='red',edgecolors='none',s=100)
    
    # 隐藏坐标轴
    plt.axes().get_xaxis().set_visible(False)
    plt.axes().get_yaxis().set_visible(False)
    plt.show()
    
    keep_running = input("go on?(y/n:)")
    if keep_running == 'n':
        break
```
