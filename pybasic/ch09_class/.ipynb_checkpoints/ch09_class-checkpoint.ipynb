## 类的创建及使用

**__init__()方法**
__init__()  是一个特殊的方法，每当你根据 Dog  类创建新实例时， Python 都会自动运行它。在这个方法的名称中，开头和末尾各有两个下划线，这是一种约定，旨在避免 Python 默认方法与普通方法发生名称冲突。

__init__()  定义成了包含三个形参： self  、 name  和 age  。在这个方法的定义中，形参 self  必不可少，还必须位于其他形参的前面。

**属性**
 self.name = name  获取存储在形参 name  中的值，并将其存储到变量 name  中，然后该变量被关联到当前创建的实例。 self.age = age  的作用与此类似。像这样可通过实例访问的变量称为属性属性   。

**py2.7中**
`
class ClassName(object):
`


````python
class Dog():
    '''
    dog class
    '''
    def __init__(self,name,age):
        '''
        初始化name 和 age的属性
        '''
        self.name = name
        self.age = age
        
    def sit(self):
        '''
        sit action
        '''
        print(self.name.title() + " is now sitting.")
        
    def roll_over(self):
        '''
        roll over
        '''
        print(self.name.title() + " rolled over!")
        
        
my_dog = Dog('gu',6)
print("My dog's name is " + my_dog.name.title() + '.')
print("My dog is " + str(my_dog.age) + " years old.")
my_dog.roll_over()
my_dog.sit()

your_dog = Dog('lucky',1)
print("\nMy dog's name is " + your_dog.name.title() + '.')
print("My dog is " + str(your_dog.age) + " years old.")
your_dog.roll_over()
your_dog.sit()
````

## 类和实例

对于属性的修改的方法：
- 直接通过实例进行修改；
- 通过方法进行设置；
- 通过方法进行递增（增加特定的值）
 

````python
class Car():
    '''Car class'''
    def __init__(self,make,model,year):
        '''init car'''
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0
        
    def get_descriptive_name(self):
        car_info = str(self.year) + ' ' + self.make +' ' + self.model
        return car_info.title()
    
    def read_odometer(self):
        print("this car has " + str(self.odometer_reading) + " miles on it.")
        
    def update_odometer(self,mileage):
        if mileage > self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("you can't roll back an odometer!")
            
    def increment_odometer(self,miles):
        if miles >0:
            self.odometer_reading += miles
        else: 
            print("you can't roll back an odometer!")
        
    
my_car = Car("bmw",'m1',2016)
print(my_car.get_descriptive_name())
my_car.read_odometer()

# 直接通过实例进行修改；
my_car.odometer_reading = 1000
my_car.read_odometer()

# 通过方法进行设置；
my_car.update_odometer(2000)
my_car.read_odometer()

my_car.update_odometer(1000)
my_car.read_odometer()

# 通过方法进行递增（增加特定的值）
my_car.increment_odometer(1000)
my_car.read_odometer()
my_car.increment_odometer(-1000)
my_car.read_odometer()
        
        
````

## 类的继承

## 继承的格式

- 类的代码（见❶）。创建子类时，父类必须包含在当前文件中，且位于子类前面。
- 在❷处，我们定义了子类 ElectricCar  。定义子类时，必须在括号内指定父类的
名称。
- 方法 __init__() 接受创建 Car  实例所需的信息（见❸）。
- ❹处的 super()  是一个特殊函数，帮助 Python 将父类和子类关联起来。

` ElectricCar(Car)`

````python
class Car():
    '''Car class'''
    def __init__(self,make,model,year):
        '''init car'''
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0
        
    def get_descriptive_name(self):
        car_info = str(self.year) + ' ' + self.make +' ' + self.model
        return car_info.title()
    
    def read_odometer(self):
        print("this car has " + str(self.odometer_reading) + " miles on it.")
        
    def update_odometer(self,mileage):
        if mileage > self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("you can't roll back an odometer!")
            
    def increment_odometer(self,miles):
        if miles >0:
            self.odometer_reading += miles
        else: 
            print("you can't roll back an odometer!")
            


class ElectricCar(Car):
    '''Electric Car class'''
    def __init__(self,make,model,year):
        super().__init__(make,model,year)
        self.battery = Battery()
        
        
class Battery():
    '''Battery class'''
    def __init__(self,battery_size = 70):
        self.battery_size = battery_size
    def describe_battery(self):
         print("this  car has a " + str(self.battery_size) + '-kWh battery.')
            
    def get_range(self):
        if self.battery_size == 70:
            range = 240
        elif self.battery_size == 85:
            range = 270
        message = "this car can go approimately " + str(range) + " miles on a full change."
        print(message)
        
        




my_tesla = ElectricCar('tesla','model S',2016)
print(my_tesla.get_descriptive_name())
my_tesla.battery.describe_battery()
my_tesla.battery.get_range()
````
